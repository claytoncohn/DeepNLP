<h1> <center>Intersection and Union of Skin Cancer Relations in Big and Little Datasets


This notebook is created for the purpose of identifying the overlap between causal relation chains in the "big" and "little" dataset partitions.

Google Drive must first be mounted to import the dataset(s).

```
from google.colab import drive
drive.mount('/content/drive')
```

Import the two datasets.

```
import pandas as pd
DATA_PATH = "drive/My Drive/colab/IJAIED/Data/"
BIG_NAME = "sc_big.tsv"
LITTLE_NAME = "sc_little.tsv"
```

```
big = pd.read_csv(DATA_PATH + BIG_NAME, delimiter='\t', header=0, names=['file', 'relation', 's_num', 'sentence'])
big.head(10)
```

```
little = pd.read_csv(DATA_PATH + LITTLE_NAME, delimiter='\t', header=0, names=['file', 'relation', 's_num', 'sentence'])
little.head(10)
```

Iterate through both partitions and catalog relations in each.

```
rels_BIG = set([])
rels_LITTLE = set([])
```

```
for index, row in big.iterrows():
  rels = row["relation"].split('|')
  for rel in rels: 
    rels_BIG.add(rel)
print("BIG: ", len(rels_BIG))
```

```
for index, row in little.iterrows():
  rels = row["relation"].split('|')
  for rel in rels: 
    rels_LITTLE.add(rel)
print("LITTLE: ", len(rels_LITTLE))
```

Isolate the intersection and disjoint sets.

```
both = rels_BIG.intersection(rels_LITTLE)
big_only = [rel for rel in rels_BIG if rel not in rels_LITTLE]
little_only = [rel for rel in rels_LITTLE if rel not in rels_BIG]
```

```
print("Intersection: ", len(both))
print("Big only: ", len(big_only))
print("Little only: ", len(little_only))
```
